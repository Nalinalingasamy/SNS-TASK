```python
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
sns.distplot(df["duration"])
```

```python
#since duration is normally distributed--- mean replacement
print(df["duration"].mean())
df["duration"] = df["duration"].fillna(df["duration"].mean())
```

```python
#data is not normal=== median replacement
print(df["num_critic_for_reviews"].median())
print(df["num_critic_for_reviews"].mode()[0])
print(df["num_critic_for_reviews"].mean())
df["num_critic_for_reviews"] = df["num_critic_for_reviews"].fillna(df["num_critic_for_reviews"].median())
```
